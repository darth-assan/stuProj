```python
# Normalizing data with MinMax
from sklearn.preprocessing import MinMaxScaler
def normalize_data(df):
    # Initialize the MinMaxScaler with default range (0, 1)
    scaler = MinMaxScaler()
    
    # Fit and transform the data, ensuring the result is a DataFrame
    normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
    
    return normalized_df
```

```python
# Dropping attack columns
clean_data = normalized_data.drop(['attack', 'attack_P1', 'attack_P2', 'attack_P3'], axis=1)
clean_data
```

```python
# Function to calculate the euclidean distances
import numpy as np
def calculate_distances(df):
    # Convert DataFrame to NumPy array for calculations
    data = df.values
    
    # Calculate pairwise differences between consecutive rows
    differences = np.diff(data, axis=0)
    
    # Calculate Euclidean distance for each pair of consecutive rows
    distances = np.sqrt(np.sum(differences ** 2, axis=1))
    
    # Create a DataFrame with row index pairs for clarity
    distance_df = pd.DataFrame({
        'Row_Pair': [f"{i}-{i+1}" for i in range(len(distances))],
        'Distance': distances
    })
    
    return distance_df
```

```python
distances = calculate_distances(clean_data)
distances.to_csv('/Users/darth/Desktop/clean_test1.csv')
```

```python
# Function to check if a column has constant values
def is_constant(df, column):
    return df[column].nunique() == 1

# Function to print columns with constant values
def print_constant_columns(df):
    constant_columns = [col for col in df.columns if is_constant(df, col)]
    for col in constant_columns:
        print(f"Column '{col}' has constant values: {df[col].unique()[0]}")

# Example usage
print_constant_columns(filtered_data)
```
